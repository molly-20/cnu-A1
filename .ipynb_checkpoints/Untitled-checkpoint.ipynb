```python
import numpy as np
import math

def simpson_I(f, a, b, N):
    h = (b-a)/N
    k = 0
    x = a + h
    for i in range(1,N//2+1):
        k += 4*f(x)
        x += 2*h

    x = a + 2*h 
    for i in range(2,(N//2)):
        k += 2*f(x)
        x += 2*h
       
    return (h/3)*(f(a)+f(b)+k)

def f(x):
    y = (x**4*(np.cos(k*x)))
    return y

Z = simpson_I(f, 0, (2*np.pi), 100)

print(f'{Z:.6f}')
    
```

```python
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt

def f(x):
    return np.cos(x)

def f_prime(x):
    return -1*np.sin(x)

def derivative_1(f, h, x):
    D1 = (f(x+h)-f(x))/h
    return D1

def derivative_2(f, h, x):
    D2 = (f(x)-f(x-h))/h 
    return D2

def derivative_3(f, h, x):
    D3 = (f(x+h)-f(x-h))/2*h
    return D3

x0 = 1
h = np.logspace(1e-16, 1e-1, 50)
d_approx1 = derivative_1(f, h, x0)
d_approx2 = derivative_2(f, h, x0)
d_approx3 = derivative_3(f, h, x0)
d_exact = f_prime(x0)


abs_diff1 = d_exact - d_approx1
print(abs_diff1)

abs_diff2 = d_exact - d_approx2
print(abs_diff2)

abs_diff3 = d_exact - d_approx3
print(abs_diff3)



    

```
